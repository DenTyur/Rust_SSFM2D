```python
import numpy as np
from matplotlib import cm
from matplotlib import pyplot as pl
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.load("/home/denis/Rust/projects/SSFM2D/src/arrays_saved/x.npy")
y = np.load("/home/denis/Rust/projects/SSFM2D/src/arrays_saved/y.npy")
u = np.load("/home/denis/Rust/projects/SSFM2D/src/arrays_saved/u.npy")
```

```python
X, Y = np.meshgrid(x, y, indexing='ij')
```

```python
fig = pl.figure(figsize=(5, 5))
ax1 = fig.add_subplot(111)
ax1.set_aspect("equal")
b1 = ax1.pcolormesh(X, Y, np.abs(u), cmap=cm.jet, shading="auto")
cb1 = pl.colorbar(b1, ax=ax1)
pl.show()
```

```python
def oscillator(X, Y):
    return np.exp(-0.5*(X**2 + Y**2))
```

```python
psi = oscillator(X, Y)
norm = np.sqrt(np.sum(np.abs(psi)**2, axis=(0,1))*0.01**2)
print(norm)
psi = psi/norm
```

```python
fig = pl.figure(figsize=(5, 5))
ax1 = fig.add_subplot(111)
ax1.set_aspect("equal")
b1 = ax1.pcolormesh(X, Y, np.abs(psi)**2, cmap=cm.jet, shading="auto")
cb1 = pl.colorbar(b1, ax=ax1)
pl.show()
```
